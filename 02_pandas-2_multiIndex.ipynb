###  .merge двух DataFrame с MultiIndex

```python
uri_1 = 'https://gist.githubusercontent.com/urwinday/67c4328e8556630c2c91f0fd8d87b174/raw/9760ec75f37e854dc6c60339abc64782e11a2cc9/Age_group'
df_Age_group_mergeMultiIndex = pd.read_csv(uri_1, index_col = ['category','type'])
df_Age_group_mergeMultiIndex.head(10)
```

```python
uri_2 = 'https://gist.githubusercontent.com/urwinday/27f014fa893b0abfcfd51c399b10a832/raw/528cfabc2cc2c689bc7bb48b846986a7776e418c/Rehabilitation'
df_Rehabilitation_mergeMultiIndex = pd.read_csv(uri_2)
df_Rehabilitation_mergeMultiIndex.head(10)
```

```python
# merge двух DF с MultiIndex
# is_state признак - государственная или не государственная
new_2_df_rehabilitated = pd.merge(df_Age_group_mergeMultiIndex, df_Rehabilitation_mergeMultiIndex)
new_2_df_rehabilitated.pivot_table(index = ['is_state','state_name'], columns='year', values='rehabilitated').fillna(0).astype(int)
```
